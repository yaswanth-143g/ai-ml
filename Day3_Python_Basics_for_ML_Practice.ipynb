### 9. If-Elif-Else

**Question:** Take a variable `marks = 72` and print the grade: A (>=90), B (>=80), C (>=70), D (>=60), F otherwise.

```python
marks = 72
if marks>=90:
    print("Grade : A") 
elif marks>=80 & marks<90:
    print("Grade : B")
elif marks>=70 & marks<80:
    print("Grade : C")
elif marks>=60 & marks<70:
    print("Grade : D")
else:
    print("Grade : F") 
```

### 10. Functions
