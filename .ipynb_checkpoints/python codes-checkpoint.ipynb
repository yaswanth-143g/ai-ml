### list comprehnsion

```python
num=[1,2,3,4,5,6]
sruares=[]

for i in num:
    sruares.append(i**2)
print(sruares)
```

```python
squares=[x**2 for x in range(1,6+1)]
print(squares)
```

```python
even=[x for x in range(20) if x%2==0]
print(even)
```

```python
words = ["python","ML","datascience"]
upper = [w.upper() for w in words]
print(upper)
```
